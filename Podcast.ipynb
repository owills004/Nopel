```python
def define_new_columns(dataset):
    # Defining new features that may be useful
    dataset['Average_Popularity_percentage'] = (dataset['Host_Popularity_percentage']+dataset['Guest_Popularity_percentage'])/2
    dataset['Interaction'] = dataset['Host_Popularity_percentage']*dataset['Guest_Popularity_percentage']
    dataset['Diff_Popularity_percentage'] = dataset['Host_Popularity_percentage'] - dataset['Guest_Popularity_percentage']
    dataset['Ad_density'] = dataset['Number_of_Ads']/dataset['Episode_Length_minutes']
    dataset['Mins_per_Ad'] = dataset['Episode_Length_minutes']/dataset['Number_of_Ads'].replace(0, np.nan)
    # Bucketize and Binning of some columns
    dataset['episode_length_bucket'] = pd.cut(dataset['Episode_Length_minutes'],
                                              bins=[0, 10, 30, 60,  np.inf],
                                             labels=['very_short', 'short', 'medium', 'long']) # BUcketizing the episode_length
    dataset['Host_bin'] = pd.cut(dataset['Host_Popularity_percentage'],
                                bins=[0, 33, 66, 100],
                                labels=['low', 'medium', 'high']) # Bin host popularity
    dataset['Guest_bin'] = pd.cut(dataset['Guest_Popularity_percentage'],
                                 bins=[0, 33, 66, 100],
                                 labels=['low', 'medium', 'high']) # Bin guest popularity

    return dataset
```

```python
# new imputers and encoders
bin_cat = [['low', 'medium', 'high'], ['low', 'medium', 'high']]
bucket = ['very_short','short', 'medium', 'long']

new_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')
bin_encoder = OrdinalEncoder(categories=bin_cat)
bucket_encoder = OrdinalEncoder(categories=[bucket])

train_df[l] = new_imputer.fit_transform(train_df[l])
val_df[l] = new_imputer.transform(val_df[l])
train_df[s] = cat_imputer.fit_transform(train_df[s])
val_df[s] = cat_imputer.transform(val_df[s])

train_df[c] = bin_encoder.fit_transform(train_df[c])
val_df[c] = bin_encoder.transform(val_df[c])

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


lr = LinearRegression()
rf = RandomForestRegressor(random_state=1)
xgb = XGBRegressor(random_state=66)

lr.fit(X_train, train_y)
rf.fit(X_train, train_y)
xgb.fit(X_train, train_y)

lr_pred = lr.predict(X_val)
rf_pred = rf.predict(X_val)
xgb_pred = xgb.predict(X_val)
```

```python
from sklearn.metrics import root_mean_squared_error

def evaluate_model(model, val_y, pred):
    print(f"Model: {model}")
    print(f"Mae: {mean_absolute_error(val_y, pred):.2f}")
    print(f"MSE: {mean_squared_error(val_y, pred):.2f}")
    print(f"RMSE: {root_mean_squared_error(val_y, pred):.2f}")
    print(f"R2 score: {r2_score(val_y, pred):.2f}")
    print("-"*30)


evaluate_model('Linear Regression', val_y, lr_pred)
evaluate_model("Random Forest", val_y, rf_pred)
evaluate_model('XGBoost', val_y, xgb_pred)
```
